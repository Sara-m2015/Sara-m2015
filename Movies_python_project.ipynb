```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df= pd.read_csv('tmdb-movies.csv')
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.describe()
```

```python
# to see if there is a duplicate
df.nunique()
```

```python
df.duplicated().sum()
```

```python
# dropping duplicares and validating the excution
df.drop_duplicates(inplace = True)
df.duplicated().any()
```

```python
# check for data frame shape
df.shape
```

```python
# dropping unnesseery column
df.drop(['id','homepage','revenue','budget','tagline','overview','keywords'], axis = 1, inplace = True)
df.shape
```

```python
df.isnull().sum()
```

```python
# Replace NaN values in 'genres' with puplic
df['genres'].fillna('puplic', inplace=True)
df.isnull().sum()
```

```python
df.info()
```

```python
# Replace NaN values in 'director' with puplic
df['director'].fillna('unkown', inplace=True)
df.isnull().sum()
```

```python
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
df.info()
```

```python
# split values in the hyprid dataframe
df['genres'] = df['genres'].apply(lambda x: x.split("|")[0] if isinstance(x, str) else x)
df.head()
```

```python
plt.figure(figsize=(10,8))
df['genres'].value_counts().plot(kind="bar")
plt.title("The most filmed genres", fontsize=(10))
plt.xlabel("genres", fontsize=10)
plt.ylabel("Count",fontsize=10)
plt.show()
```

```python
# adding new column for movie profit (revenue_adj - budget_adj)
df['profit_adj'] = df.revenue_adj - df.budget_adj
df.head()
```

```python
sns.scatterplot(x='budget_adj', y='profit_adj', data=df)
plt.title('Relationship between budget and profit after accounting for inflation')
plt.show()
```

```python
df['genres'] = df['genres'].astype(str)
plt.scatter(df['genres'], df['profit_adj'])
plt.xlabel('genres')
plt.ylabel('profit_adj')
plt.title('Relationship between genres and profit after accounting for inflation')
# Rotate x-axis labels for better readability
plt.xticks(rotation=90)
plt.show()
```

```python
sns.lineplot(x='vote_average', y='popularity', data=df)
plt.title('Relationship between vote_average and popularity')
plt.show()
```

```python
sns.scatterplot(x='vote_average', y='popularity', data=df)
plt.title('Relationship between vote_average and popularity')
plt.show()
```

```python
genres_means = df.groupby('genres').agg({'budget_adj': 'mean', 'profit_adj': 'mean'})
fig, ax1 = plt.subplots(figsize=(15, 10))

# Plot mean budget using seaborn
sns.lineplot(x=genres_means.index, y=genres_means['budget_adj'], ax=ax1, color='blue', marker='o', label='Mean Budget')
ax1.set_xlabel('genres')
ax1.set_ylabel('Mean Budget', color='blue')
ax1.tick_params('y', labelcolor='blue')

# Create the second y-axis
ax2 = ax1.twinx()

# Plot mean profit using seaborn
sns.lineplot(x=genres_means.index, y=genres_means['profit_adj'], ax=ax2, color='red', marker='s', label='Mean Profit')
ax2.set_ylabel('Mean Profit', color='red')
ax2.tick_params('y', labelcolor='red')

# Add legends
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')

plt.title('Mean Budget and Profit by Genres')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
df['month_name'] = df['release_date'].dt.month_name()
df.head()
```

```python
sns.lineplot(x='month_name', y='popularity', data=df)
plt.title('Relationship between Month and popularity')
plt.xticks(rotation=90)
plt.show()
```

```python
# split values in the hyprid dataframe
df['director'] = df['director'].apply(lambda x: x.split("|")[0] if isinstance(x, str) else x)
df.head()
```

```python
# Group by director and calculate mean profit
director_profit = df.groupby('director')['profit_adj'].mean().sort_values(ascending=False).head(10)
fig, ax1 = plt.subplots(figsize=(15, 10))
sns.lineplot(x=director_profit.index, y=director_profit.values, marker='o')
plt.xlabel('director')
plt.ylabel('profit_adj')
plt.title('Relationship between Top 10 director and profit after accounting for inflation')
plt.xticks(rotation=90)
plt.show()
```

```python
director_profit = df.groupby('director')['profit_adj'].mean().sort_values(ascending=False).head(10)
fig, ax1 = plt.subplots(figsize=(15, 10))
sns.lineplot(x=director_profit.index, y=director_profit.values, marker='o')
ax1.set_xlabel('director')
ax1.set_ylabel('profit_adj')
ax1.set_title('Relationship between Top 10 director, profit and budjet after accounting for inflation')
ax1.set_xticks(range(len(director_profit.index)))
ax1.set_xticklabels(director_profit.index, rotation=90)
ax2 = ax1.twinx()
director_budget = df.groupby('director')['budget_adj'].mean().sort_values(ascending=False).head(10)
sns.lineplot(x=director_budget.index, y=director_budget.values, ax=ax2, color='red', marker='s')
ax2.set_ylabel('budget_adj', color='red')
ax2.tick_params('y', labelcolor='red')
plt.show()
```

```python
df['production_companies'].fillna('unkown', inplace=True)
df.isnull().sum()
```

```python
df['production_companies'] = df['production_companies'].apply(lambda x: x.split("|")[0] if isinstance(x, str) else x)
# Group by director and calculate mean profit
production_companies_profit = df.groupby('production_companies')['profit_adj'].mean().sort_values(ascending=False).head(10)
fig, ax1 = plt.subplots(figsize=(15, 10))
sns.lineplot(x=production_companies_profit.index, y=production_companies_profit.values, marker='o')
plt.xlabel('production_companies')
plt.ylabel('profit_adj')
plt.title('Relationship between Top 10 production companies and profit after accounting for inflation')
plt.xticks(rotation=90)
plt.show()
```
